```python
import numpy as np
import scipy
from scipy.special import elliprd
from scipy.special import elliprf

import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.colors as colors
import sqrtscale
plt.style.use('matplotlibrc')
with plt.style.context('matplotlibrc'):
    plt.show()
```

```python
def H(x):
    return np.where(x < 0, 0, 1)
```

```python
def a(s):
    if(s<h):
        return 0
    else:
        return np.sqrt(1-((h*h)/(s*s)))
```

```python
def m(s,d):
    if(s < d):
        return (s*s)/(d*d)
    else:
        return (d*d)/(s*s)
```

```python
def R_A(f,x):
    if(x>=-1 and x<=1):
        return x*np.sqrt(1 - (x*x)) * f - np.arcsin(x)
    else:
        print('arcsin error')
```

```python
def b(s):
    if(s<d):
        return s/d
    else:
        return 1
```

```python
def R_x(u,m):
    z1,z2,z3 = 1., 1.-(u*u), 1.-(m*u*u)
    rdz = elliprd(z1,z2,z3)
    rfz = elliprf(z1,z2,z3)
    R_x = np.sqrt((1.-(u*u))*(1.-(m*u*u))) + ((2./3.)*(m+1)*u*u*rdz) - rfz
    return R_x
```

```python
def c1(s):
    if(s<d):
        #print('a(s): ',a(s))
        return(((8.*h) / (np.pi * (d + (2.*h)) * s)) * (((1/3.) * R_x(1, m(s,d))) - ((a(s)/3.) * R_x(a(s), m(s,d)))))   
    else:
        #print('s*a(s)/d: ',s*a(s)/d)
        return(((8.*h*d) / (np.pi * (d + (2.*h)) * s * s)) * (((1/3.) * R_x(1, m(s,d))) - ((s*a(s)/d/3.) * R_x(s*a(s)/d, m(s,d)))))
```

```python
def c02(s):
    return (d / (np.pi * (d + (2.*h)) * (s**3)) * ( ((d*d) * (R_A((6.*b(s)*b(s))+1. , b(s)) - R_A((6.*(s*a(s)/d)*(s*a(s)/d))+1 , (s*a(s)/d)) ) ) - ((4.*h*h) * H(s-h) * (R_A(-1,1) - R_A(-1,(s*a(s)/d))) ) ) )
    
```

```python
h=2./np.sqrt(5.)
d=4./np.sqrt(5.)
#h=np.sqrt(2)
#d=np.sqrt(2)

c1s = []
c02s = []
for s in np.arange(0.001,2,0.001):
    #print("s: ",s)
    #print("c1(s): ",c1(s))
    #print("c02(s): ",c02(s))
    c1s.append(c1(s))
    c02s.append(c02(s))

plt.clf()
#plt.scatter(np.arange(0.1,2,0.01),c02s)
#plt.scatter(np.arange(0.1,4,0.01),c1s)    
plt.xlim(0,2)
plt.ylim(-1,5)
plt.scatter(np.arange(0.001,2,0.001),np.add(np.array(c1s),np.array(c02s)),label='total')
plt.scatter(np.arange(0.001,2,0.001),np.array(c1s),label='c1s')
plt.scatter(np.arange(0.001,2,0.001),np.array(c02s),label='c02s')
#plt.yscale('log')
plt.grid(which='major')
plt.grid(which='minor')
plt.legend()
plt.show()
```
